```python
# Import Library
```

```python
import pandas as pd
```

```python
# Open the Data set
```

```python
dataset=pd.read_csv("50_Statups.csv")
```

```python
dataset
```

```python
# Remove the categorical data(i.e str-state)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
# splite the input & output columns
```

```python
dataset.columns
```

```python
independant=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
independant
```

```python
dependant= dataset[["Profit"]]
```

```python
dependant
```

```python
#split the Train & Test dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
# Model Improve for using the Standardisation
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_test
```

```python
# Train data-Model creation
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor(criterion='squared_error',splitter='best')
regressor=regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# Evaluation Method Presidcting the Test set result
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score =r2_score(Y_test,Y_pred)
```
